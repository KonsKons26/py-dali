```python
import sys
import os


sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
print(sys.path[-1])
```

```python
from src.utils import quick_load
from src.monte_carlo import MonteCarloAligner, ProteinAlignment
from src.scores import substructure_similarity_score
```

```python
BASE_DIR = os.path.dirname(os.getcwd())
REDUCED_DIRS = os.path.join(BASE_DIR, "reduced")

print(f"{'BASE_DIR':<15}{BASE_DIR}")
print(f"{'REDUCED_DIRS':<15}{REDUCED_DIRS}")
```

```python
sample1 = quick_load(
    os.path.join(REDUCED_DIRS, "1euv_A.pkl.gz")
)

sample2 = quick_load(
    os.path.join(REDUCED_DIRS, "1tgz_A.pkl.gz")
)
```

```python
self_similarity1 = sum([
    substructure_similarity_score(
        sample1[i][-1],
        sample1[i][-1]
    )
    for i in range(len(sample1))
])

aligner = MonteCarloAligner(sample1, sample1)

# Force diagonal alignment
diagonal_pairs = [(i, i) for i in range(len(sample1))]
diagonal_alignment = ProteinAlignment(pairs=diagonal_pairs)
diagonal_score = diagonal_alignment.score(
    aligner.contact_patterns_A, 
    aligner.contact_patterns_B
)

print(f"Diagonal score equals direct sum: {0.9999 < diagonal_score/self_similarity1 < 1.00001}")
```

```python
mca_s1_to_s1 = MonteCarloAligner(
    sample1, sample1,
    iteration_limit=1_000,
    betas_mode="linear"
)
mca_s1_to_s1.run_simulation(verbose=True)
mca_s1_to_s1.plot_convergence()
mca_s1_to_s1.best_alignment
```

```python
self_similarity2 = sum([
    substructure_similarity_score(
        sample2[i][-1],
        sample2[i][-1]
    )
    for i in range(len(sample2))
])

aligner = MonteCarloAligner(sample2, sample2)

# Force diagonal alignment
diagonal_pairs = [(i, i) for i in range(len(sample2))]
diagonal_alignment = ProteinAlignment(pairs=diagonal_pairs)
diagonal_score = diagonal_alignment.score(
    aligner.contact_patterns_A, 
    aligner.contact_patterns_B
)

print(
    "Diagonal score equals direct sum: "
    f"{0.9999 < diagonal_score/self_similarity2 < 1.00001}"
)
```

```python
mca_s2_to_s2 = MonteCarloAligner(
    sample2, sample2,
    iteration_limit=1_000,
    betas_range=(0.1, 100.0)
    betas_mode="linear"
)
mca_s2_to_s2.run_simulation(verbose=True)
mca_s2_to_s2.plot_convergence()
mca_s2_to_s2.best_alignment
```

```python
mca_s1_to_s2 = MonteCarloAligner(
    sample1, sample2,
    iteration_limit=10_000,
    betas_mode="exponential"
)
mca_s1_to_s2.run_simulation(verbose=True)
mca_s1_to_s2.plot_convergence()
mca_s1_to_s2.best_alignment
```
