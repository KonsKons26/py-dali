```python
import sys
import os


sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
print(sys.path[-1])
```

```python
BASE_DIR = os.path.dirname(os.getcwd())
META_DIR = os.path.join(BASE_DIR, "meta")
DATA_DIR = os.path.join(BASE_DIR, "data")
QUERIES_META_DIR = os.path.join(META_DIR, "queries")

os.makedirs(DATA_DIR, exist_ok=True)
os.makedirs(QUERIES_META_DIR, exist_ok=True)

print(f"{'BASE_DIR':<20}{BASE_DIR}")
print(f"{'META_DIR':<20}{META_DIR}")
print(f"{'DATA_DIR':<20}{DATA_DIR}")
print(f"{'QUERIES_META_DIR':<20}{QUERIES_META_DIR}")
```

```python
import pandas as pd
import random

from src.utils import select_domains, download_and_process_domains
```

```python
df = pd.read_csv(
    os.path.join(META_DIR, "astral.csv")
)
df
```

```python
seed = 66
n = 5
m = 5

rng = random.Random(seed)
```

```python
# Select domains
domains = select_domains(df, n, m, rng)
for query, reference in domains.items():
    print(f"Query: {query}, References: {reference["domain_id"].tolist()}")
    reference.to_csv(
        os.path.join(QUERIES_META_DIR, f"{query}.csv"),
        index=False
    )
    print(f"Saved {query} metadata to {os.path.join(QUERIES_META_DIR, f'{query}.csv')}")
    print()
```

```python
# Download and process domains
download_and_process_domains(domains, DATA_DIR)
```
